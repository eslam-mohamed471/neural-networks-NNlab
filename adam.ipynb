```python
import numpy as np
import tensorflow as tf
```

```python
N, D, H= 64, 1000, 100
x = tf.placeholder(tf.float32, shape=(N, D))
y = tf.placeholder(tf.float32, shape=(N, D))
model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(H, input_shape=(D,), activation=tf.nn.relu))
model.add(tf.keras.layers.Dense(D))
model.compile(loss=tf.keras.losses.mean_squared_error, optimizer=tf.keras.optimizers.Adam(lr=1e0))
```

```python
x = np.random.randn(N, D)
y = np.random.randn(N, D)
history = model.fit(x, y, epochs=500, batch_size=N)
```
